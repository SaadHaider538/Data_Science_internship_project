```python

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download NLTK resources (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize tools
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    # 1. Tokenize
    tokens = word_tokenize(text.lower())

    # 2. Remove stopwords and non-alphanumeric tokens
    tokens = [word for word in tokens if word.isalnum() and word not in stop_words]

    # 3. Lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return " ".join(tokens)
```

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load dataset
df = pd.read_csv('creditcard.csv')
print(df.head())

# Step 3: Split features and target
X = df.drop('Class', axis=1)
y = df['Class']

# Check for null values and drop them if necessary
if X.isnull().sum().sum() > 0:
    print("Null values found. Dropping rows with missing values...")
    df = df.dropna()
    X = df.drop('Class', axis=1)
    y = df['Class']

# Ensure all columns are numeric
X = X.select_dtypes(include=[np.number])

# Now apply SMOTE
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X, y)

print(f"Before SMOTE: {np.bincount(y)}")
print(f"After SMOTE: {np.bincount(y_resampled)}")

# Step 5: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Step 6: Train Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Step 7: Evaluate model
y_pred = model.predict(X_test)
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))

# Step 8: Confusion matrix
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Step 9: Simple Command-Line Interface to Test the Model
def test_transaction_input():
    print("\n--- Fraud Detection System ---")
    input_features = []
    for i in range(1, 29):  # V1 to V28
        val = float(input(f"Enter value for V{i}: "))
        input_features.append(val)
    amount = float(input("Enter transaction amount: "))
    input_features.append(amount)

    input_df = pd.DataFrame([input_features], columns=X.columns)
    prediction = model.predict(input_df)[0]

    if prediction == 1:
        print("  Fraudulent Transaction Detected!")
    else:
        print(" Transaction appears legitimate.")

# Uncomment to use CLI:
# test_transaction_input()
```

Predicting House Prices Using the Boston Housing Dataset

 Build a regression model from scratch to predict house prices using the Boston Housing Dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Load Boston Housing Dataset
boston = pd.read_csv('housing.csv')
X = boston.drop('MEDV', axis=1)
y = boston['MEDV']


```

```python
# Normalize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
class LinearRegression:
    def __init__(self):
        self.coef_ = None

    def fit(self, X, y):
        # Add a column of ones for the intercept term
        X = np.c_[np.ones(X.shape[0]), X]
        # Calculating the coefficients using the Normal Equation
        self.coef_ = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

    def predict(self, X):
        # Add a column of ones for the intercept term
        X = np.c_[np.ones(X.shape[0]), X]
        return X.dot(self.coef_)

# Train the model
linear_reg = LinearRegression()
linear_reg.fit(X_scaled, y)
y_pred_lr = linear_reg.predict(X_scaled)

# Evaluate the model
rmse_lr = np.sqrt(mean_squared_error(y, y_pred_lr))
r2_lr = r2_score(y, y_pred_lr)
print(f"Linear Regression - RMSE: {rmse_lr:.3f}, R: {r2_lr:.3f}")
```

Implementing Random Forest from Scratch

```python
class DecisionTree:
    def __init__(self, max_depth=None):
        self.max_depth = max_depth
        self.tree = None

    def fit(self, X, y):
        self.tree = self._build_tree(X, y)

    def _build_tree(self, X, y, depth=0):
        n_samples, n_features = X.shape
        best_split = None
        best_score = float('inf')

        if depth >= self.max_depth:
            return np.mean(y)

        # Try every feature and every possible split value to find the best split
        for feature_index in range(n_features):
            possible_values = np.unique(X[:, feature_index])
            for value in possible_values:
                left_mask = X[:, feature_index] <= value
                right_mask = ~left_mask

                left_y = y[left_mask]
                right_y = y[right_mask]

                # Calculate the variance (simplified criterion for regression)
                score = self._calculate_variance(left_y, right_y)

                if score < best_score:
                    best_score = score
                    best_split = {
                        'feature': feature_index,
                        'value': value,
                        'left': left_mask,
                        'right': right_mask
                    }

        if best_split:
            left_y = y[best_split['left']]
            right_y = y[best_split['right']]
            left_tree = self._build_tree(X[best_split['left']], left_y, depth+1)
            right_tree = self._build_tree(X[best_split['right']], right_y, depth+1)
            return {'split': best_split, 'left': left_tree, 'right': right_tree}
        return np.mean(y)

    def _calculate_variance(self, left_y, right_y):
        left_var = np.var(left_y)
        right_var = np.var(right_y)
        return left_var * len(left_y) + right_var * len(right_y)

    def predict(self, X):
        return np.array([self._predict_sample(x, self.tree) for x in X])

    def _predict_sample(self, x, tree):
        if isinstance(tree, dict):
            split = tree['split']
            if x[split['feature']] <= split['value']:
                return self._predict_sample(x, tree['left'])
            else:
                return self._predict_sample(x, tree['right'])
        return tree

# Random Forest with multiple decision trees
class RandomForest:
    def __init__(self, n_trees=10, max_depth=5):
        self.n_trees = n_trees
        self.max_depth = max_depth
        self.trees = []

    def fit(self, X, y):
        for _ in range(self.n_trees):
            # Bootstrap sampling
            indices = np.random.choice(range(X.shape[0]), size=X.shape[0], replace=True)
            X_sample = X[indices]
            y_sample = y[indices]
            tree = DecisionTree(max_depth=self.max_depth)
            tree.fit(X_sample, y_sample)
            self.trees.append(tree)

    def predict(self, X):
        tree_preds = np.array([tree.predict(X) for tree in self.trees])
        return np.mean(tree_preds, axis=0)

# Train Random Forest model
rf_model = RandomForest(n_trees=10, max_depth=5)
rf_model.fit(X_scaled, y)
y_pred_rf = rf_model.predict(X_scaled)

# Evaluate Random Forest model
rmse_rf = np.sqrt(mean_squared_error(y, y_pred_rf))
r2_rf = r2_score(y, y_pred_rf)
print(f"Random Forest - RMSE: {rmse_rf:.3f}, R: {r2_rf:.3f}")
```
